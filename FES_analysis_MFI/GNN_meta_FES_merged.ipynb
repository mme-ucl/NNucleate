```python
analytical_x=np.concatenate([analytical_x_WT3[:2*NH3-1],analytical_x_WT1[:2*NH1-1],analytical_x_WT2[:2*NH2-1],analytical_x_METAD[:2*NH0-1]])
analytical_y=np.concatenate([analytical_y_WT3[:2*NH3-1],analytical_y_WT1[:2*NH1-1],analytical_y_WT2[:2*NH2-1],analytical_y_METAD[:2*NH0-1]])
weights=np.concatenate([NH3*weightsWT3,NH1*weightsWT1,NH1*weightsWT2,NH0*weightsMETAD])

position_x=np.concatenate([position_x_WT3[:2*NH3-1],position_x_WT1[:2*NH1-1],position_x_WT2[:2*NH2-1],position_x_METAD[:2*NH0-1]])
position_y=np.concatenate([position_y_WT3[:2*NH3-1],position_y_WT1[:2*NH1-1],position_y_WT2[:2*NH2-1],position_y_METAD[:2*NH0-1]])



#Reweight the probability density in the analytical CV space and compute reweighted F(S)
[Prewe,Xrewe,Yrewe]=np.histogram2d(analytical_x**(1/3), analytical_y**(1/3),bins=50,weights=weights)
Prewe = Prewe.T
Frewe = -kT*np.log(Prewe)-np.min(-kT*np.log(Prewe))
```

```python
#convergence analysis
start=0

fig, axs = plt.subplots(5,2, figsize=(10,25))

for block in range(5):
    stop=int(start+37900)
    indexes=np.arange(0,stop,1,dtype=int)
    [Prewe,Xrewe,Yrewe]=np.histogram2d(position_x[indexes], position_y[indexes],bins=50,weights=weights[indexes])
    Prewe = Prewe.T
    Frewe = -kT*np.log(Prewe)-np.min(-kT*np.log(Prewe))
    
    plot0=axs[block,0].contourf(Xrewe[:-1], Yrewe[:-1],Frewe,cmap='Spectral',levels=np.arange(0,50, 1), antialiased=False, alpha=0.8)
    
    axs[block,0].set_ylabel('model $\sqrt[3]{n(Q6)}$',fontsize=20)
    axs[block,0].set_xlabel('model $\sqrt[3]{n}$',fontsize=20)
    axs[block,0].set_xlim(np.min(Xrewe),np.max(Xrewe))
    axs[block,0].set_ylim(np.min(Yrewe),np.max(Yrewe))
    cbar = plt.colorbar(plot0, ax=axs[block,0])
    cbar.set_label("$\Delta F^*$",fontsize=20)
    axs[block,0].set_xlim(np.min(position_x),np.max(position_x))
    axs[block,0].set_ylim(np.min(position_y),np.max(position_y))
 
    [Prewe,Xrewe,Yrewe]=np.histogram2d(analytical_x[indexes]**(1/3), analytical_y[indexes]**(1/3),bins=25,weights=weights[indexes])
    Prewe = Prewe.T
    Frewe = -kT*np.log(Prewe)-np.min(-kT*np.log(Prewe))
    plot1=axs[block,1].contourf(Xrewe[:-1], Yrewe[:-1],Frewe,cmap='Spectral',levels=np.arange(0, 50, 1), antialiased=False, alpha=0.8)
    axs[block,1].set_ylabel('analytical $\sqrt[3]{n(Q6)}$',fontsize=20)
    axs[block,1].set_xlabel('analytical $\sqrt[3]{n}$',fontsize=20)

    axs[block,1].set_xlim(np.min(Xrewe),np.max(Xrewe))
    axs[block,1].set_ylim(np.min(Yrewe),np.max(Yrewe))
    cbar = plt.colorbar(plot1, ax=axs[block,1])
    cbar.set_label("$\Delta F^*$",fontsize=20)
    axs[block,1].set_xlim(np.min(analytical_x**(1/3)),np.max(analytical_x**(1/3)))
    axs[block,1].set_ylim(np.min(analytical_y**(1/3)),np.max(analytical_y**(1/3)))
   
    
    start=stop
    
plt.savefig('convergence_ALL', dpi=600, bbox_inches='tight')
    
    
```
